Pie Chart
Pie Chart digunakan untuk menyajikan data perbandingan dalam bentuk presentase. Cocok untuk mengetahui nilai peluang kemunculan data.

```python
import matplotlib.pyplot as plt
nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
Jumlah_permen = [52, 39, 78, 13, 78]
Warna = ['#E3F2FD','#8D6E63','#4FC3F7','#EC407A','#FFAB91']
Highlight = (0,0.2,0,0,0)

plt.title('Peluang Permen\n')
plt.pie(Jumlah_permen,labels = nama_permen, autopct = '%1.1f%%', colors = Warna, explode = Highlight)
plt.show()
```

# Soal 3: Makanan

Bar Chart
Bar Chart adalah visualisasi data dalam bentuk batang yang bertujuan untuk mencari hasil olah data antar dua variable yang saling berhubungan.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

Makanan_penutup = ('Donat','Pastel','Kue Coklat','Ice Cream','Puding Vanila',
                   'Brownies','Puding Strawberi','Puding Coklat','Ice Cream Nutela',
                    'Kue Coklat-Keju','Kue Wajik','Kue Sus','Mochi')
banyak = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({'Penjualan': Makanan_penutup, 'Qty': banyak})
df.sort_values(by='Qty', inplace = True, ascending = False)

kor_x = np.arange(len(df))
warna = ['#1976D2' for _ in range(len(df))]
warna [-3] = '#F44336'
warna [-2] = '#F44336'
warna [-1] = '#F44336'

plt.title('Menu Makanan Kafe Biru\n')
plt.bar(kor_x, df['Qty'],tick_label=df['Penjualan'],color = warna)
plt.xticks(rotation=90)
plt.ylabel('Banyak Makanan\n')
plt.xlabel('\nDaftar Menu Makanan Penutup')
plt.show()
```

Dari hasil diagram batang(Bar Chart) diatas yang memiliki popularitas tertinggi adalah Ice Cream dan yang terendah adalah Kue Wajik. 
Tiga makanan penutup yang harus disingkirkan adalah Puding Vanila, Pastel dan Kue Wajik.

Bar Chart sangat membantu dalam melihat grafik penjualan makanan penutup Kafe Biru dengan baik dan mempermudah mengurutkan data dari tertinggi sampai terendah atau sebaliknya
